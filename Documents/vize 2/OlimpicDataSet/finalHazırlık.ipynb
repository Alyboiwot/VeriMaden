```python
import pandas as pd
import numpy as np

# Veriyi biz tanÄ±mlayalÄ±m (4 satÄ±r, 3 sÃ¼tun)
data = [[8, 6, 10],   # KiÅŸi A
        [10, 8, 6],   # KiÅŸi B
        [5, 9, 10],   # KiÅŸi C
        [12, 7, 5]]   # KiÅŸi D

# Etiketlerimiz
rows = ["A", "B", "C", "D"]
colum = ["WORKHOUR", "SLEEPHOUR", "RESTTIME"]

# DataFrame'i oluÅŸturalÄ±m
df = pd.DataFrame(data, index=rows, columns=colum)

print("--- ManipÃ¼lasyon iÃ§in Temel DataFrame ---")
print(df)
print("\n" + "-"*40 + "\n")
```

## 2. Filtreleme (KoÅŸullu SeÃ§im) ðŸŽ¯

```python
# Ã‡alÄ±ÅŸma saati (WORKHOUR) 8'den fazla olanlar
print("--- Ã‡alÄ±ÅŸma Saati > 8 Olanlar ---")
print(df[df['WORKHOUR'] > 8])
# Ã‡Ä±ktÄ±:
#    WORKHOUR  SLEEPHOUR  RESTTIME
# B        10          8         6
# D        12          7         5
print("\n")

# Uyku saati (SLEEPHOUR) 7'den az olanlar
print("--- Uyku Saati < 7 Olanlar ---")
print(df[df['SLEEPHOUR'] < 7])
# Ã‡Ä±ktÄ±:
#    WORKHOUR  SLEEPHOUR  RESTTIME
# A         8          6        10
print("\n")

# Ä°KÄ° KOÅžUL BÄ°RDEN ( & = 've' )
# Ã‡alÄ±ÅŸma saati 5'ten fazla VE Uyku saati 8'den az olanlar
print("--- WORKHOUR > 5 VE SLEEPHOUR < 8 Olanlar ---")
print(df[(df['WORKHOUR'] > 5) & (df['SLEEPHOUR'] < 8)])
# Ã‡Ä±ktÄ±:
#    WORKHOUR  SLEEPHOUR  RESTTIME
# A         8          6        10
# D        12          7         5
```

## 2. merge DurumlarÄ± (SQL JOIN Tipleri)

# merge ise "eÅŸleÅŸen veri" olmadan Ã§alÄ±ÅŸamaz. on='anahtar_sÃ¼tun parametresine ihtiyaÃ§ duyar. "Merge durumlarÄ±" (how parametresi) ise, bu anahtar sÃ¼tundaki deÄŸerler eÅŸleÅŸmediÄŸinde ne yapÄ±lacaÄŸÄ±nÄ± belirler.

# Finansal Kod Ã–rneÄŸi: Bitcoin ve Nasdaq fiyatlarÄ±nÄ± 'Tarih' sÃ¼tununa gÃ¶re birleÅŸtirmek istiyorsunuz. Ancak hafta sonlarÄ± Kripto (btc) aÃ§Ä±kken Nasdaq (qqq) kapalÄ±dÄ±r (yani tarihler tam eÅŸleÅŸmez).

```python
# Bitcoin verisi (Hafta sonu dahil)
btc = pd.DataFrame({
    'Tarih': ['2024-01-05', '2024-01-06', '2024-01-07'],
    'BTC_Fiyat': [43000, 44000, 43500]
})

# Nasdaq verisi (Hafta sonu kapalÄ±)
qqq = pd.DataFrame({
    'Tarih': ['2024-01-05', '2024-01-08'],
    'QQQ_Fiyat': [400, 402]
})

# Ortak eÅŸleÅŸen tek gÃ¼n: '2024-01-05'
```

## a) how='inner' (Ä°Ã§ BirleÅŸtirme - VarsayÄ±lan)

#### Sadece her iki tabloda da 'Tarih' eÅŸleÅŸmesi bulunan satÄ±rlarÄ± getirir

```python
pd.merge(btc, qqq, on='Tarih', how='inner')
```

## b) how='left' (Sol BirleÅŸtirme)

#### Soldaki (ilk yazÄ±lan, btc) tablodaki tÃ¼m tarihleri getirir. SaÄŸdaki (qqq) tabloda eÅŸleÅŸen veri varsa yazar, yoksa NaN koyar.

```python
pd.merge(btc, qqq, on='Tarih', how='left')
```

## c) how='right' (SaÄŸ BirleÅŸtirme)

##### SaÄŸdaki (ikinci yazÄ±lan, qqq) tablodaki tÃ¼m tarihleri getirir. Soldaki (btc) tabloda eÅŸleÅŸen veri varsa yazar, yoksa NaN koyar.

```python
pd.merge(btc, qqq, on='Tarih', how='right')
```

## d) how='outer' (DÄ±ÅŸ BirleÅŸtirme)

 #### Her iki tablodaki tÃ¼m tarihleri getirir. EÅŸleÅŸmeyen yerleri NaN ile doldurur.

pd.merge(btc, qqq, on='Tarih', how='outer')

## .apply() metodu
Pandas'taki .apply() metodu, bir DataFrame'in her bir satÄ±rÄ± veya sÃ¼tunu Ã¼zerinde karmaÅŸÄ±k, Ã¶zel fonksiyonlarÄ± Ã§alÄ±ÅŸtÄ±rmak iÃ§in kullanÄ±lan en esnek ve gÃ¼Ã§lÃ¼ araÃ§lardan biridir.

BasitÃ§e sÃ¶ylemek gerekirse, .apply() bir dÃ¶ngÃ¼ yazmanÄ±za gerek kalmadan her satÄ±r veya sÃ¼tun iÃ§in bir fonksiyonu "uygular".

ðŸ’¡ .apply() Ne Zaman KullanÄ±lÄ±r?

df['A'] + df['B'] gibi basit matematiksel iÅŸlemlerin yetersiz kaldÄ±ÄŸÄ± durumlarda .apply() kullanÄ±lÄ±r. Ã–zellikle if-else mantÄ±ÄŸÄ±na dayalÄ± yeni bir sÃ¼tun oluÅŸturmak istediÄŸinizde bu metoda ihtiyaÃ§ duyarsÄ±nÄ±z.

âš™ï¸ NasÄ±l Ã‡alÄ±ÅŸÄ±r: axis Parametresi

.apply() metodunun davranÄ±ÅŸÄ±nÄ± belirleyen en Ã¶nemli parametre axis'tir:

axis=0 (VarsayÄ±lan): Fonksiyonu her bir sÃ¼tuna uygular. (SÃ¼tun bazlÄ± iÅŸlem).

axis=1: Fonksiyonu her bir satÄ±ra uygular. (SatÄ±r bazlÄ± iÅŸlem).

En yaygÄ±n ve kullanÄ±ÅŸlÄ± olanÄ± axis=1'dir.


## Ã–rnek 1: SatÄ±r BazlÄ± (axis=1) - Finansal Strateji SÃ¼tunu OluÅŸturma

Senaryo: Bitcoin (BTC_Getiri) ve Nasdaq (QQQ_Getiri) gÃ¼nlÃ¼k getirilerini iÃ§eren bir DataFrame'iniz var. Bu iki veriye bakarak bir "Sinyal" sÃ¼tunu oluÅŸturmak istiyorsunuz:

EÄŸer ikisi de pozitifse "GÃ¼Ã§lÃ¼ Al"

EÄŸer BTC pozitif, QQQ negatifse "BTC Al"

DiÄŸer tÃ¼m durumlarda "Bekle"


```python
import pandas as pd
import numpy as np

# 1. Ã–rnek Veri Seti
df = pd.DataFrame({
    'BTC_Getiri': [0.02, -0.01, 0.03, -0.02],
    'QQQ_Getiri': [0.01, 0.01, -0.01, -0.03]
})

print("--- ORÄ°JÄ°NAL VERÄ° ---")
print(df)

# 2. Ã–zel Fonksiyonumuzu TanÄ±mlama
# Bu fonksiyon tek bir 'row' (satÄ±r) alÄ±r
def sinyal_uret(row):
    btc = row['BTC_Getiri']
    qqq = row['QQQ_Getiri']
    
    if btc > 0 and qqq > 0:
        return "GÃ¼Ã§lÃ¼ Al"
    elif btc > 0 and qqq <= 0:
        return "BTC Al"
    else:
        return "Bekle"

# 3. .apply() KullanÄ±mÄ±
# axis=1: Pandas'a 'sinyal_uret' fonksiyonunu her bir satÄ±r iÃ§in ayrÄ± ayrÄ± Ã§alÄ±ÅŸtÄ±r
df['Sinyal'] = df.apply(sinyal_uret, axis=1)

print("\n--- .apply() SONRASI YENÄ° SÃœTUN ---")
print(df)
```

## Ã–rnek 2: SÃ¼tun BazlÄ± (axis=0) - Her VarlÄ±k Ä°Ã§in Ã–zet Ä°statistik

Senaryo: AynÄ± DataFrame'deki her bir sÃ¼tun (BTC_Getiri ve QQQ_Getiri) iÃ§in maksimum getiriyi ve minimum getiriyi hesaplayÄ±p bir Ã¶zet Ã§Ä±karmak istiyorsunuz

```python
# 1. Ã–rnek Veri (YukarÄ±dakiyle aynÄ±)
df = pd.DataFrame({
    'BTC_Getiri': [0.02, -0.01, 0.03, -0.02],
    'QQQ_Getiri': [0.01, 0.01, -0.01, -0.03]
})

# 2. Ã–zel Fonksiyon (Bu fonksiyon tek bir 'column' (sÃ¼tun) alÄ±r)
def sutun_ozeti(column):
    return f"Maks: {column.max():.2f}, Min: {column.min():.2f}"

# 3. .apply() KullanÄ±mÄ±
# axis=0: 'sutun_ozeti' fonksiyonunu Ã¶nce 'BTC_Getiri' sÃ¼tunu, 
# sonra 'QQQ_Getiri' sÃ¼tunu iÃ§in Ã§alÄ±ÅŸtÄ±r.
ozet = df.apply(sutun_ozeti, axis=0)

print("\n--- SÃœTUN BAZLI Ã–ZET (axis=0) ---")
print(ozet)
```

## Ã–rnek 3: lambda ile HÄ±zlÄ± KullanÄ±m (axis=1)

EÄŸer fonksiyonunuz Ã§ok kÄ±saysa (tek satÄ±rlÄ±k if-else gibi), def ile ayrÄ± bir fonksiyon tanÄ±mlamak yerine lambda kullanabilirsiniz.

Senaryo: Bitcoin ve Nasdaq arasÄ±ndaki getiri farkÄ±nÄ± ('Spread') hesaplayan yeni bir sÃ¼tun oluÅŸturun.

```python
# 'lambda' kullanarak ayrÄ± bir fonksiyon tanÄ±mlamaya gerek kalmaz
df['Spread'] = df.apply(lambda row: row['BTC_Getiri'] - row['QQQ_Getiri'], axis=1)

print("\n--- LAMBDA Ä°LE YENÄ° SÃœTUN ---")
print(df)
```

GÃ¶rev: DataFrame'e Bonus adÄ±nda yeni bir sÃ¼tun ekle. Bonus sÃ¼tunu, Department (Departman) sÃ¼tununa gÃ¶re Salary (MaaÅŸ) Ã¼zerinden hesaplansÄ±n:


            Department == "Sales" (SatÄ±ÅŸ) ise â†’ MaaÅŸÄ±n %10'u (Salary * 0.10)

             Department == "Marketing" (Pazarlama) ise â†’ MaaÅŸÄ±n %8'i (Salary * 0.08)

              Department == "Finance" (Finans) ise â†’ MaaÅŸÄ±n %5'i (Salary * 0.05)

               DiÄŸer tÃ¼m departmanlar â†’ 0 (SÄ±fÄ±r bonus)

```python
 def BonusForDept (row):
    Department = row["Department"]
    sal = row["Salary"]
    bonus = 0
    if Department == "Sales":
         bonus = sal*0.10
       
    elif  Department == "Marketing":
         bonus = sal*0.08
     
    elif  Department == "Finance" :
         bonus = sal*0.05
    return bonus   
   
dfMyEmp["BONUSLABEL"]= dfMyEmp.apply(BonusForDept,axis=1)   
dfMyEmp
        
    
```

```python
import pandas as pd
from scipy.stats import shapiro

# Veriyi yÃ¼kle
df = pd.read_csv("WineQT.csv")

# Test edilecek sÃ¼tunlar
columns_to_test = ['alcohol', 'pH', 'density']

print(f"{'DEÄžÄ°ÅžKEN':<15} | {'P-DEÄžERÄ°':<10} | {'SONUÃ‡'}")
print("-" * 45)

for col in columns_to_test:
    # Shapiro-Wilk Testi
    stat, p_value = shapiro(df[col])
    
    # Sonucu Yorumlama
    if p_value > 0.05:
        sonuc = "NORMAL DAÄžILIYOR (H0 Reddedilemez)"
    else:
        sonuc = "NORMAL DEÄžÄ°L (H0 Reddedildi)"
        
    print(f"{col:<15} | {p_value:.5f}    | {sonuc}")

print("-" * 45)
print("* Not: P-deÄŸeri 0.05'ten kÃ¼Ã§Ã¼kse normal daÄŸÄ±lÄ±m varsayÄ±mÄ± bozulur.")
```
