````python
import numpy as np 
import pandas as pd
import seaborn as sns
````

````python
np.random.seed(42)

set1no = 900
set2no = 100
````

````python
## dataframe with imbalanced data set
df1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=set1no),
    'feature_2': np.random.normal(loc=0, scale=1, size=set1no),
    'target': [0] * set1no
})

df2 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=set2no),
    'feature_2': np.random.normal(loc=2, scale=1, size=set2no),
    'target': [1] * set2no
})
````

````python
sns.boxplot(data=df1)
````

````python
df=pd.concat([df1,df2]).reset_index(drop=True)
````

````python
df
````

````python
df["target"].unique()
````

````python
df["target"].value_counts()
````

## Resampling

````python
df_majority = df[df["target"] == 0]
df_minority = df[df["target"] == 1]
````

````python
df_majority
````

````python
from sklearn.utils import resample
````

````python
df_minority_upsampled = resample(df_minority,replace=True,n_samples=len(df_majority),random_state=42)
````

````python
df_minority_upsampled
````

````python
df_Upsampled= pd.concat([df_minority_upsampled,df_majority])
````

````python
df_Upsampled
````

````python
df_majority_DownSampled = resample(df_majority,replace=True,n_samples=len(df_minority),random_state=42)
````

````python
df_majority_DownSampled 
````

````python
pd.concat([df_majority_DownSampled,df_minority])
````

SMOTE (Synthetic Minority Oversampling Technique - Sentetik Azınlık Fazla Örnekleme Tekniği), makine öğrenmesinde **sınıf dengesizliği (Class Imbalance)** sorununu çözmek için kullanılan en popüler ve etkili yöntemdir.

Bu teknik, azınlık sınıfındaki örnek sayısını yapay olarak artırarak modelin bu nadir durumları görmezden gelmesini engellemeyi amaçlar.

-----

### 1\. Temel Tanım ve Amaç

  * **Çözdüğü Sorun:** Sınıflama verilerinde, bir sınıfın (Örn: Dolandırıcılık, Hastalık, Nadir Kripto İşlemi) diğerinden (%99) çok daha az örneğe sahip olması durumunda, model bu %1'lik azınlık sınıfını öğrenmek yerine görmezden gelmeyi öğrenir.
  * **Amaç:** Modeli eğitirken her iki sınıfa da eşit sayıda örnek sunmak, böylece modelin azınlık sınıftaki önemli desenleri kaçırmamasını sağlamaktır.

### 2\. SMOTE Nasıl Çalışır? (Sentetik Üretim)

SMOTE'un gücü, azınlık sınıfındaki mevcut örnekleri **kopyalamaması** değil, **yeni, yapay (sentetik) örnekler üretmesidir**.

1.  **Örnek Seçimi:** Azınlık sınıftan rastgele bir veri noktası ($P$) seçilir.
2.  **Komşu Bulma:** Bu noktaya ($P$) en yakın olan komşular ($K$) bulunur.
3.  **Sentez:** SMOTE, seçilen nokta ($P$) ile komşularından biri arasında düz bir çizgi çizer. Bu çizgi üzerinde **rastgele bir noktaya** yeni, yapay bir örnek ($S$) oluşturulur.

Bu süreç defalarca tekrarlanarak azınlık sınıfının sayısı çoğunluk sınıfla eşitlenir. Bu sayede modeliniz **aynı veriyi tekrar tekrar görmek** yerine, azınlık sınıfın etrafında biraz daha farklılık içeren **yapay ama mantıklı** veriler görmüş olur.

### 3\. Kodda Uygulanışı

Bu işlem, genellikle `imblearn` kütüphanesindeki `fit_resample` fonksiyonu ile yapılır:

```python
from imblearn.over_sampling import SMOTE

# SMOTE nesnesini oluştur
sm = SMOTE(random_state=42)

# Uygula ve Dengeli Yeni Veriyi Al
X_resampled, y_resampled = sm.fit_resample(X, y)
```

  * `fit_resample` komutu, azınlık sınıftan yeterli sayıda yapay örnek üreterek **toplam veri setinin boyutunu büyütür**.
  * `X_resampled` ve `y_resampled` artık dengelenmiş, model eğitimine hazır yeni özellikler ve etiketlerdir.
