```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Understanding Box Plots
# Box plots (also called whisker plots) are used to visualize the distribution of a dataset.
# They provide insights into:
# - The median (Q2, 50th percentile)
# - The interquartile range (IQR) - spread between Q1 (25th percentile) and Q3 (75th percentile)
# - The minimum and maximum values within 1.5 * IQR (whiskers)
# - Outliers (values outside the whiskers)
```

```python
data = np.array([5, 7, 9, 15, 20, 22, 25, 30, 32, 35, 37, 40, 50, 55, 60, 100])

plt.figure(figsize=(6, 5))
sns.boxplot(y=data)
plt.title("Box Plot of Simple Numeric Data")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
### Interpretation ###
# 1. The box represents the interquartile range (IQR), which contains 50% of the data.
# 2. The line inside the box is the median (50th percentile).
# 3. The whiskers extend to 1.5 times the IQR.
# 4. Any points beyond the whiskers are considered outliers.
# 5. Higher class passengers generally paid higher fares, but there are extreme outliers (very high fares).
```

```python
data = np.array([5, 7, 9, 15, 20, 22, 25, 30, 32, 35, 37, 40, 50, 55, 60,100,500])

plt.figure(figsize=(6, 5))
sns.boxplot(y=data)
plt.title("Box Plot of Simple Numeric Data")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
df = sns.load_dataset("titanic")
```

```python
df
```

```python
plt.figure(figsize=(15, 5))

# Box plot for Age grouped by Class
plt.subplot(1, 2, 1)
sns.boxplot(x="class", y="age", data=df)
plt.title("Box Plot of Age by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.grid(True)

# Box plot for Fare grouped by Class
plt.subplot(1, 2, 2)
sns.boxplot(x="class", y="fare", data=df)
plt.title("Box Plot of Fare by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.yscale("log")  # Log scale for better visualization of high fares
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Higher class passengers generally paid higher fares, but there are extreme outliers (very high fares).
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Rastgelelik iÃ§in tohum (hep aynÄ± sonucu almak iÃ§in)
np.random.seed(42)

# --- Veri SimÃ¼lasyonu ---
# Senaryo 1: BoÄŸa PiyasasÄ± (Bull Market) - Ã–rn: 2021
# GÃ¼nlÃ¼k ortalama getiri pozitif (%0.1), volatilite dÃ¼ÅŸÃ¼k (%1)
bull_returns = np.random.normal(loc=0.1, scale=1.0, size=250)

# Senaryo 2: AyÄ± PiyasasÄ± (Bear Market) - Ã–rn: 2022
# GÃ¼nlÃ¼k ortalama getiri negatif (-%0.05), volatilite yÃ¼ksek (%2) ve ÅŸok dÃ¼ÅŸÃ¼ÅŸler var
bear_returns = np.random.normal(loc=-0.05, scale=2.0, size=250)
# AyÄ± piyasasÄ±na birkaÃ§ "Ã§Ã¶kÃ¼ÅŸ" (extreme outlier) ekleyelim
bear_returns = np.append(bear_returns, [-5.5, -6.2, -4.8, 4.5]) 

# DataFrame oluÅŸturma
df_bull = pd.DataFrame({'Getiri': bull_returns, 'Piyasa': 'BoÄŸa YÄ±lÄ± (Bull)'})
df_bear = pd.DataFrame({'Getiri': bear_returns, 'Piyasa': 'AyÄ± YÄ±lÄ± (Bear)'})
df_nasdaq = pd.concat([df_bull, df_bear])

# --- GÃ¶rselleÅŸtirme ---
plt.figure(figsize=(10, 6))
sns.boxplot(x='Piyasa', y='Getiri', data=df_nasdaq, palette="Set2")

plt.title('Nasdaq GÃ¼nlÃ¼k Getiri DaÄŸÄ±lÄ±mÄ±: BoÄŸa vs. AyÄ± PiyasasÄ±', fontsize=14)
plt.ylabel('GÃ¼nlÃ¼k Getiri (%)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.axhline(0, color='red', linestyle='-', linewidth=1) # 0 Ã§izgisi (KÃ¢r/Zarar sÄ±nÄ±rÄ±)
plt.show()
```
